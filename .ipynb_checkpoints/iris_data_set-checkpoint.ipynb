```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Importa as bibliotecas necessárias para criar a IA
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# treino AI
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

rf_reg = RandomForestRegressor()
rf_reg.fit(X_train, y_train)
```

```python
# teste AI
test_pred_lin = lin_reg.predict(X_test)
test_pred_rf = rf_reg.predict(X_test)

# Calcula as metricas (R^2 e erro quadrado médio)
# Para saber qual o melhor modelo
r2_lin = metrics.r2_score(y_test, test_pred_lin)
mse_lin = metrics.mean_squared_error(y_test, test_pred_lin)

print(f"R² da Regressão Linear: {r2_lin}")
print(f"MSE da Regressão Linear: {mse_lin}")
r2_rf= metrics.r2_score(y_test, test_pred_rf)
mse_rf = metrics.mean_squared_error(y_test, test_pred_rf)
print(f"R² do Random Forest: {r2_rf}")
print(f"MSE do Random Forest: {mse_rf}")
```

### Procure o que tem o maior R^2 e o menor MSE
1) Em outras palavras, o que tem o maior grau de alinhamento (R^2) e o menor erro (MSE)

2) Nesse caso, Random Forest se saiu melhor!

```python
# Arredonda valores da previsão
display(test_pred_rf)
aux = []
for i in range(len(test_pred_rf)):
    aux.append(round(test_pred_rf[i]))
    
aux = np.array(aux)
#aux = round(test_pred_rf)
display(aux)
test_pred_rf = aux
```

```python
# Arredonda valores da previsão
aux = []
for i in range(len(test_pred_lin)):
    aux.append(round(test_pred_lin[i]))
aux = np.array(aux)
display(aux)
test_pred_lin = aux
display(test_pred_lin)
```

```python
# Cria dataframe pra guardar os dados
df_resultado = pd.DataFrame()
# df_resultado.index = x_test

# Guarda os y_teste
df_resultado['y_teste'] = y_test

# Predição da random forest
df_resultado['y_previsao_rf'] = test_pred_rf

# Predição linear
df_resultado['y_previsao_lin'] = test_pred_lin

display(df_resultado)
df_resultado = df_resultado.reset_index(drop=True)
plt.figure(figsize=(15, 5))
sns.lineplot(data=df_resultado)
plt.show()
display(df_resultado)
```
